# Plotting choropleth maps with python

```python
import cartopy.io.shapereader as shpreader
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
### read shapefile from naturalearth to geopandas DataFrame (GeoDataFrame)

shapename = shpreader.natural_earth(resolution='50m', category='cultural', name='admin_0_countries')
countries = gpd.read_file(shapename)
```

```python
### in principle, GeoDataFrames behave like standard pandas DataFrames:

### you can select subsets
europe = countries.loc[countries.index[countries.CONTINENT=='Europe'],:]

### drop entries
europe.drop(europe.index[europe.NAME=='Russia'], inplace=True)

### add entries
europe['fundamentally important variable'] = 0.

### and plot
europe.plot('POP_EST', legend=True)
plt.xlim(-20,41)
plt.ylim(35,75)
plt.show()

europe.plot('fundamentally important variable', legend=True)
plt.xlim(-20,41)
plt.ylim(35,75)
plt.show()

### plus same other fancy stuff using the geoemtry attribute (the shape information)
europe.dissolve('CONTINENT').plot('POP_EST', legend=True)
plt.xlim(-20,41)
plt.ylim(35,75)
plt.show()
```
